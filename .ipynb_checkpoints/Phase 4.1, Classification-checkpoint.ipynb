### Gaussian Naive Bayes

```python
GNB = GaussianNB(var_smoothing= 0.8111308307896871)
tscv = TimeSeriesSplit(max_train_size=None, n_splits=len(X)-1)

print(tscv)
GNBpred = []
GNBerror = []
i = 0
for train_index, test_index in tscv.split(X.values):
    if len(train_index) < lenTrain:
        continue
    Xtrain, Xtest = X.loc[train_index], X.loc[test_index]
    ytrain, ytest = y.loc[train_index], y.loc[test_index]
    
    #scaller
    Xtrain = scaler.fit_transform(Xtrain)
    Xtest = scaler.transform(Xtest)
    
    #fit the model
    GNB.fit(Xtrain, ytrain)
    y_pred = GNB.predict(Xtest)
#     print(y_test , y_pred,  y_test- y_pred)
    GNBpred.append(y_pred[0])
    GNBerror.append(ytest - y_pred[0])
    i += 1

```

```python
MethodNameList.append('Gaussian Naive Bayes')
f1score_weightedList.append(0.62)
MSEList.append(0.316)
```

```python
# Evaluation
print('GNB accuracy', metrics.accuracy_score(y_test, GNBpred))
print(classification_report(y_test, GNBpred))

print('GNB MSE', mean([np.power(er, 2) for er in GNBerror]))
figure(figsize=(16, 12))

plt.plot([er[0] for er in GNBerror])
```

### Linear Discriminant Analysis

```python
lda = LinearDiscriminantAnalysis(shrinkage=0.01, solver='lsqr')

tscv = TimeSeriesSplit(max_train_size=None, n_splits=len(X)-1)

print(tscv)
ldapred = []
ldaerror = []
i = 0
for train_index, test_index in tscv.split(X.values):
    if len(train_index) < lenTrain:
        continue
    Xtrain, Xtest = X.loc[train_index], X.loc[test_index]
    ytrain, ytest = y.loc[train_index], y.loc[test_index]
    
    #scaller
    Xtrain = scaler.fit_transform(Xtrain)
    Xtest = scaler.transform(Xtest)
    
    #fit the model
    lda.fit(Xtrain, ytrain)
    y_pred = lda.predict(Xtest)
#     print(y_test , y_pred,  y_test- y_pred)
    ldapred.append(y_pred[0])
    ldaerror.append(ytest - y_pred[0])
    i += 1

```

```python
MethodNameList.append('Linear Discriminant Analysis')
f1score_weightedList.append(0.68)
MSEList.append(0.291)
```
