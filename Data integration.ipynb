```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
pollution = pd.read_excel('Pollution_Restelo_2019_2020.xlsx')
pollutionBufDate = []
for i in range(len(pollution)):
    pollutionBufDate.append(pollution.Restelo[i].strftime("%Y-%m-%d"))
pollution.Restelo = pollutionBufDate
pollution
```

```python
weatherData = pd.read_excel('Lisbon,Portugal 2019-01-01 to 2020-03-31.xlsx')
weatherData = weatherData.merge(pollution, left_on='datetime',
                                right_on='Restelo', how='inner').drop('Restelo',axis=1)
```

```python
pollenBuf = pd.read_excel('Pollen_2019.xlsx').iloc[:,:2]
pollenBufDate = []
for i in range(len(pollenBuf)):
    pollenBufDate.append(pollenBuf.date[i].strftime("%Y-%m-%d"))
pollenBuf.date = pollenBufDate
pollenBuf.date[0]
# pollenBufDate[0]
```

```python
pollenData = pd.DataFrame()
pollenData['date'] = weatherData['datetime']
# wheatherData['datetime'] = pollenData['date']
pollenData['pollen_intensity'] = 'good'
pollenData
```

```python
intensityBuf = pollenBuf['pollen_intensity']
datepollenBuf = list(pollenBuf['date'])
pollen_intensity = pollenData['pollen_intensity']
datepollenData = pollenData['date']

for i in range(len(pollen_intensity)):
    if datepollenData[i] in datepollenBuf:
#         print(datepollenBuf[datepollenBuf.index(datepollenData[i])],
#               datepollenData[i],
#               intensityBuf[datepollenBuf.index(datepollenData[i])])
        pollen_intensity[i] = intensityBuf[datepollenBuf.index(datepollenData[i])]
pollenData['pollen_intensity'] = pollen_intensity
```

```python
admisionData = pd.read_excel('URG_ASMA_LAST.XLSX')
admisionData
```

```python
admisionDate = []
for i in range(len(admisionData)):
#     try:
        admisionDate.append(admisionData.data_admissao[i].strftime("%Y-%m-%d"))
#     except:
#         print(i)
admisionData.data_admissao = admisionDate
admisionData.data_admissao
```

```python
# admisionData.data_admissao[0].strftime("%Y-%m-%d")
```

```python
asmaData = admisionData.merge(pollenData, left_on='data_admissao', right_on='date', how='inner').\
merge(weatherData, left_on='data_admissao', right_on='datetime', how='inner')
asmaData.to_csv('integrated Asma Data.csv', index=False)
```

```python
weather = weatherData.merge(pollenData, left_on='datetime', right_on='date', how='inner').\
drop(['date'], axis=1)
```

```python
weather['admission number'] = 0
dumy_df = pd.DataFrame()
# print()
dumy_df['datetime'] = list(admisionData.groupby(['data_admissao']).size().index)
dumy_df['athmaCounter'] = list(admisionData.groupby(['data_admissao']).size())
dumy_df
```

```python
for i in range(len(weather)):
    for j in range(len(dumy_df)):
        if weather.loc[i, 'datetime'] == dumy_df.loc[j, 'datetime']:
            weather.loc[i, 'admission number'] = dumy_df.loc[j, 'athmaCounter']
weather['admission number']
```

```python
weather.to_csv('everyday weather and admissions.csv', index=False)
```
